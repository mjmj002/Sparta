portfolio = {

2. **현재가 조회 함수 작성 (`get_current_prices_api`)**

   - 응답을 JSON으로 변환 후, 코인별 현재가(`trade_price`)를 딕셔너리에 담아 반환한다.

- **포트폴리오 정의**  
보유 코인을 딕셔너리 형태로 저장합니다.

  }

- **현재가 조회**  
업비트 API를 호출해서 받은 JSON 데이터에서 `trade_price` 값을 꺼내면 됩니다.
  ```python
  for data in prices_data:

- **가치 계산**  
각 코인의 가치는 `현재가 × 수량`으로 구하고, 총합은 반복문에서 누적합니다.
  ```python
  value = price * quantity
  total_value += value

- **비중 계산**  
총합이 구해졌다면, 각 항목에 비중을 추가할 수 있습니다.
  ```python
  item["비중"] = (item["가치"] / total_value) * 100

````python
import requests
from datetime import datetime
import time
import json

# 문제 1의 답안을 작성해주세요
# 현재가 조회 함수 작성 (get_current_prices_api)
def get_current_prices_api(tickers):

    url = "https://api.upbit.com/v1/ticker" # url 을 통해 ticker 모든 암포화폐 불러옴
    headers = {"accept": "application/json"} # 딕셔너리 형태

    markets_param = ",".join(tickers)  # 쉼표 기준으로 tickers를 붙여서 넣음
    params = {"markets": markets_param}

    response = requests.get(url, headers=headers, params=params)
    prices_data = response.json() # 응답 데이터

    # print(f"prices_data : {prices_data}") # 디버깅 위한 출력문(문제판단)

    prices = {} # 가격 담는 딕셔너리
    for data in prices_data:
      prices[data['market']] = data['trade_price'] # 데이터 중 trade_price 지금 거래되고 있는 데이터만 딕셔너리에 넣음

    return prices

#prices = get_current_prices_api(["KRW-BTC"])
#print(prices)

# 포트폴리오 분석 함수
def analyze_portfolio(portfolio):

    # 1. 현재가 조회
    tickers = list(portfolio.keys()) # 암호화폐 이름(키값)만 불러옴
    # print(f"tickers : {tickers}")
    prices = get_current_prices_api(tickers)
    # print(f"prices : {prices}")
    portfolio_analysis = []
    total_value = 0

    # 분석
    for ticker, quantity in portfolio.items():
      # print(f"ticker : {ticker}, quantity : {quantity}")
      current_price = prices.get(ticker, 0 )
      # print(f"current_price : {current_price}")

      value = current_price * quantity # 현재가 * 개수 = 나의 총 자산
      # print(f"value = {value}")

      coin_name = ticker.split("-")[1]
      # print(f"coin_name = {coin_name}")

      portfolio_analysis.append({
          "코인" : coin_name,
          "수량" : quantity,
          "현재가" : current_price,
          "가치" : value
      })

      total_value += value
    # 각 코인 정보 출력
    for item in portfolio_analysis:
      item["비중"] = (item["가치"] / total_value) * 100

    print("="*95)
    print("포트폴리오 분석 결과")
    print("="*95)
    print(f"\n총 포트폴리오 가치: {total_value:,.0f}원\n")
    print("="*95)
    print(f"{'코인':^10} {'수량':>15} {'현재가':>18} {'가치':>18} {'비중':>12}")
    print("="*95)

    for item in portfolio_analysis:
      coin = item["코인"]
      quantity = item["수량"]
      price = item["현재가"]
      value = item["가치"]
      ratio = item["비중"]

      #print(f"{coin:^10} {quantity:>15.4f} {price:>18,} {value:>18,.0f} {ratio:>11.2f}%")
      print(f"{coin:^12} {quantity:>18.4f} {price:>22,} {value:>20,.0f} {ratio:>12.2f}%")

      print("-"*95)

# 사용 예시
portfolio = {
    "KRW-BTC": 0.1,
    "KRW-ETH": 3,
    "KRW-XRP": 1000
}
analyze_portfolio(portfolio)
````

- **반복 확인**  
  - 여러 번 가격을 확인해야 하므로 **반복문**을 사용합니다.
  - 각 단계에서 현재 시간과 함께 현재가를 출력하면 더 직관적입니다.

````python
# 문제 2의 답안을 작성해주세요
def get_single_price_api(ticker):
    url = "https://api.upbit.com/v1/ticker"
    headers = {"accept": "application/json"}

    params = {"markets": ticker}

    response = requests.get(url, headers=headers, params=params)
    data = response.json()[0]
    return data['trade_price']

def price_alert_system(ticker, target_high_price, target_low_price, check_count=30, interval=30):
    coin_name = ticker.split("-")[1]  # ticker의 이름만 불러옴
    print("="*80)
    print(f"{coin_name} 가격 알림 시스템 시작")
    print("="*80)
    print(f"상한가 목표: {target_high_price:,.0f}원")
    print(f"하한가 목표: {target_low_price:,.0f}원")
    print(f"확인 횟수: {check_count}회")
    print(f"확인 간격: {interval}초")
    print("-"*80)
    print()

    for i in range(check_count):
      current_price = get_single_price_api(ticker) # ticker하나만 타겟으로 한개만 불러옴

      current_time = datetime.now().strftime("%H:%M:%S")
      print(f"[{current_time}] 현재가 : {current_price:,.0f}원", end=" ")

      if current_price >= target_high_price:  # 현재가 > 상한가
        print(f"상한가 목표 금액 도달 (목표 : {target_high_price:,.0f}원")

      elif current_price <= target_low_price:
        print(f"하한가 청산 목표 금액 도달 (목표 : {target_low_price:,.0f}원)")

      else:
        print("아직 내가 설정한 목표 금액 안에 있습니다!")

      if i < check_count - 1:
        time.sleep(interval) # 초 단위아닌 interval로 체크

    print()
    print("-"*80)
    print("모니터링 종료")
    print("="*80)

ticker = "KRW-BTC"
current_price = get_single_price_api(ticker)

target_high_price = current_price * 1.05
target_low_price = current_price * 0.95

print(f"상한가 : {target_high_price}, 하한가 :{target_low_price}, 현재가 : {current_price}")
price_alert_system(ticker, target_high_price=target_high_price, target_low_price=target_low_price, check_count = 30, interval=2)
````

2. **과거 데이터 조회 함수 작성 (`get_historical_data_api`)**

   - 응답 데이터에서 투자 시점의 종가(`trade_price`)를 추출한다.

````python
import requests
from datetime import datetime, timedelta

# 과거 데이터 조회 함수
def get_historical_data_api(ticker, days_count):
  url = "https://api.upbit.com/v1/candles/days"
  params = {"market": ticker, "count": days_count}
  response = requests.get(url, params=params)
  candle_data = response.json()
  buy_price = candle_data[-1]['trade_price']
  return buy_price

# 현재가 조회 함수
def get_single_price_api(ticker):
  url = "https://api.upbit.com/v1/ticker"
  params = {"markets": ticker}
  response = requests.get(url, params=params)
  data = response.json()[0]
  return data['trade_price']

# 투자 수익률 계산 함수
def calculate_investment_return(ticker, days_count, investment_amount):
  # 과거 현재 데이터 가져옴
  buy_price = get_historical_data_api(ticker, days_count)
  current_price = get_single_price_api(ticker)

  # 투자 시점 날짜 계산
  investment_date_past = datetime.now() - timedelta(days=days_count)
  investment_date = investment_date_past.strftime('%Y-%m-%d')

  # 수익성 계산
  # (1) 투자 시점 가격 확인 : 투자 시점 가격으로 투자 금액 ÷ 가격 = 구매 수량
  quantity = investment_amount / buy_price
  # (2) 현재 가치 계산 : 구매 수량 × 현재가 = 현재 가치
  current_value = quantity * current_price
  # (3) 손익 및 수익률 계산 :
  # 손익 = 현재 가치 – 투자 금액
  profit_loss = current_value - investment_amount
  # 수익률 = (손익 ÷ 투자 금액) × 100
  yield_rate = (profit_loss / investment_amount) * 100

  # 결과 출력
  # 투자 일자, 투자 금액, 투자 시점 가격, 구매 수량, 현재 가치, 손익, 수익률을 차례대로 출력
  coin_name = ticker.split("-")[1]

  print("=" * 30)
  print(f"{coin_name} 암호화폐 수익률 계산기")
  print("=" * 30)
  print(f"투자 일자 : {investment_date}")
  print(f"투자 금액 : {investment_amount:,.0f}원")
  print(f"투자 시점 가격 : {buy_price:,.0f}원")
  print(f"구매 수량 : {quantity:,.0f}개")
  print(f"현재 가치 : {current_price:,.0f}원")
  print(f"손익 : {profit_loss:,.0f}원")
  print(f"수익률 : {yield_rate:,.1f}%")
  print("=" * 30)
  print()

# 사용 예시
ticker = "KRW-BTC"
days_count = 30
investment_amount = 1000000000

calculate_investment_return(ticker, days_count, investment_amount)
````
